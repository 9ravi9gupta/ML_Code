# Logistic Regression by Scratch

```python
import pandas as pd
import numpy as np
import math
```

```python
Uncleaned_data = pd.read_csv('data.xls')
print('shape of input data',Uncleaned_data.shape)
```

```python
Uncleaned_data.head(3)
```

```python
#    1. Replacing 'na' with 'nan' for droping further
#    2. Droping perform
#    3. Removing id column

Uncleaned_data.replace(to_replace='na',value=np.nan,inplace=True)
Uncleaned_data.dropna(axis=1,inplace=True)
Uncleaned_data = Uncleaned_data.iloc[:, 1:]
Uncleaned_data.head(3)
```

```python
Uncleaned_data['F0'] = np.ones((Uncleaned_data.shape[0],1)) #----adding one extra columns which will be used as theta_zero multipicant in sigmoid function
# Uncleaned_data.head(108)
```

# Dataset Normalization 

Here normalisation process is implemented on dataset to compress the variations and make the calculation of scores(i.e power_of_e) easy and efficient

```python
columns_mean = np.array(Uncleaned_data.mean(axis=0))  #--finding mean of each columns ,,return (31,1) ndArray

# print(columns_mean.shape)

columns_mean =columns_mean.reshape(1,columns_mean.shape[0]) #---converting columns_mean to (1,31) ndArray

Normalized_array = pd.DataFrame(np.array(Uncleaned_data.iloc[:,1:])/columns_mean)  #---dividing each columns value by their respective means

Normalized_array['diagnosis'] = Uncleaned_data['diagnosis']  #--adding diagnosis columns in new normalized dataframe

Normalized_array.head()
```

```python
#--separating the benign class rows---

BenignTumorData = pd.DataFrame(Normalized_array[Normalized_array['diagnosis']=='B'].iloc[:,0:Normalized_array.shape[1]-1])
BenignTumorData.shape
```

```python
#--separating the malignant class rows---

MalignantTumorData = pd.DataFrame(Normalized_array[Normalized_array['diagnosis']=='M'].iloc[:,0:Normalized_array.shape[1]-1])
MalignantTumorData.shape
```


1.BenignTumorData = (317,31) 

2.MalignantTumorData = (212,31)

3.Theta = (1,31)

4.power_of_e_Benign = (317,1)

5.power_of_e_malignant = (212,1)


```python
Sigma_f = lambda x: 1/(1+x)

def sigmoid_function(old_theta):
    Theta = old_theta
    
    #print("Theta",Theta,'\n')
    
    #----Here finding power of exponent of sigmoid function(sigma(theta))
    
    power_of_e_benign = ((np.array(BenignTumorData)*Theta).sum(axis=1)).reshape(BenignTumorData.shape[0],1)

    #print("power_of_e_benign",power_of_e_benign)
    
    power_of_e_malignant = ((np.array(MalignantTumorData)*Theta).sum(axis=1)).reshape(MalignantTumorData.shape[0],1)
    
    #print("\n power_of_e_malignant",power_of_e_malignant)
   
    #-------then finding sigmoid function value
    
    sigma_function_benign = pd.DataFrame(np.exp(-power_of_e_benign))[0].apply(Sigma_f)  #--call Sigma_f() function for each element of Series 
    #print(sigma_function_benign)
    sigma_function_malignant = pd.DataFrame(np.exp(-power_of_e_malignant))[0].apply(Sigma_f)  #-- #--call Sigma_f() function for each element of Series
     
    
    return(sigma_function_benign,sigma_function_malignant)  #--returning sigma_function Series
```

## Calculating Derivative of negative log liklihood function

```python
def diff_matrix(old_theta):
    sigma_function_benign , sigma_function_malignant = sigmoid_function(old_theta)
    l=[]
    for feature in BenignTumorData.columns:
        l.append((BenignTumorData[feature]*(1-sigma_function_benign)).sum(axis=0) - (MalignantTumorData[feature]*(sigma_function_malignant)).sum(axis=0))

    return(np.array(l).reshape(BenignTumorData.shape[1],1))
```

## Calculating negative log liklihood function

```python
f2 = lambda x: math.log(x)

def liklihood_fun(old_theta):
    O_T = old_theta
    sigma_function_benign , sigma_function_malignant = sigmoid_function(O_T)
#     print("function",sigma_function_benign)
    
    
    log_of_benign_fun = sigma_function_benign.apply(f2)   #--calling f2() for each element of Series
    
    log_of_malignant_fun = sigma_function_malignant.apply(f2)  #-- #--calling f2() for each element of Series
    
    liklihood_function = -(log_of_benign_fun.sum(axis=0) + log_of_malignant_fun.sum(axis=0))
    
    return(liklihood_function)
    
```

# Apply Gradient Descent Algo To Find Best Theta's 

```python
learning_rate = 0.0001
epsilon = 0.001
old_theta = np.random.rand(BenignTumorData.shape[1],1)   #--taking any random theta's as a parameters
i=0

while(True):
    
    OLD_Neg_log_fun = liklihood_fun(old_theta.T)
    
        
    new_theta = old_theta + learning_rate*diff_matrix(old_theta.T)    
    
    NEW_Neg_log_fun = liklihood_fun(new_theta.T)
    
    print(OLD_Neg_log_fun , NEW_Neg_log_fun)
    
    if ((NEW_Neg_log_fun - OLD_Neg_log_fun) < epsilon) or (i>=1100):
        break
    
    print(i)
    old_theta = new_theta
    i+=1
```

# Testing Our Logistic Regression Model

```python
f = lambda x:1/float(1+np.exp(-x))   #--Sigmoid Function--

row_index = np.random.randint(low=0,high=Uncleaned_data.shape[0])  
print(row_index)

Testing_row = np.array(Normalized_array.iloc[row_index , 0:Normalized_array.shape[1]-1 ]) #--selecting any random row 

# print(Testing_row.shape)
# print(old_theta.shape)

power_of_e = (Testing_row.reshape(Testing_row.shape[0],1) * old_theta).sum(axis=0)

# print(power_of_e[0])

posterior_prob_benign = f(power_of_e[0])

posterior_prob_malignant = 1-posterior_prob_benign

if posterior_prob_benign > posterior_prob_malignant:
    print("Tumor is Benign and posterior_prob_benign:%.2f" %(posterior_prob_benign*100))
else:
    print("Tumor is malignant and posterior_prob_malignant:%.2f" %(posterior_prob_malignant*100))
```
